4. Nobel ödülünü en çok kazanan ülkenin hangi yıldan itibaren hakimiyet sağladığını görselleştirip bu hakimiyette rol oynayan şeyler nelerdir? İçgörülerinizi paylaşır mısınız?



```python
df['birth_country'].value_counts()
```

```python
country_counts = df['birth_country'].value_counts()
top_country = country_counts.idxmax()
```

```python
top_country_df = df[df['birth_country'] == top_country]
top_country_yearly = top_country_df.groupby('year').size().cumsum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(top_country_yearly.index, top_country_yearly.values, marker='o', linestyle='-', color='b')
plt.title(f'{top_country} ülkesinin yıllara göre Nobel Ödülü kazanmaları')
plt.xlabel('Yıl')
plt.ylabel('Kümülatif Ödül Sayısı')
plt.grid(True)
plt.show()
```

5. Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')
df['year'] = pd.to_datetime(df['year'], errors='coerce')


df['age_at_award'] = df['year'].dt.year - df['birth_date'].dt.year


```

```python
categories = df['category'].unique()

for category in categories:
    plt.figure(figsize=(14, 6))

    
    plt.subplot(1, 2, 1)
    cat_df = df[df['category'] == category]
    sns.histplot(data=cat_df, x='age_at_award', hue='sex', multiple='dodge', kde=True, palette='viridis')
    plt.title(f'Age Distribution of Nobel Prize Winners in {category}')
    plt.xlabel('Age')
    plt.ylabel('Count')

   
    plt.subplot(1, 2, 2)
    sns.boxplot(data=cat_df, x='sex', y='age_at_award', palette='viridis')
    plt.title(f'Age Distribution by Gender in {category}')
    plt.xlabel('sex')
    plt.ylabel('Age at Award')

    plt.tight_layout()
    plt.show()

```

6. 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.

```python
df=data = pd.read_csv("/Users/apple/Downloads/proje.csv")
df['year'] = pd.to_datetime(df['year'], format='%Y', errors='coerce').dt.year

# Belirli yıllar arasında filtreleme yapalım
df_filtered = df[(df['year'] >= 1938) & (df['year'] <= 1945)]

# Kategorilere göre ayrı grafikler oluşturalım
categories = df_filtered['category'].unique()

for category in categories:
    plt.figure(figsize=(8, 6))
    category_data = df_filtered[df_filtered['category'] == category]
    sns.countplot(data=category_data, x='birth_country', palette='Set2')
    plt.title(f'Nobel Ödülü Kazanan Ülkeler ({category}) - 1938-1945')
    plt.xlabel('Ülke')
    plt.ylabel('Kazanan Sayısı')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()





```

8. Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.


```python
df['year'] = pd.to_datetime(df['year'], format='%Y', errors='coerce').dt.year


# Seçilen kategoriler listesi
selected_categories = ["Chemistry", "Literature", "Peace", "Physics", "Medicine"]

# Belirli yıllara ve kategorilere göre filtrelenmiş veri
df_filtered = df[(df['year'] >= 2000) & (df['category'].isin(selected_categories))]

# 'birth_date' sütunundaki eksik değerleri çıkartalım
df_filtered = df_filtered.dropna(subset=["birth_date"])

# 'birth_date' sütununu datetime türüne dönüştürelim
df_filtered["birth_date"] = pd.to_datetime(df_filtered["birth_date"], errors="coerce")

# Ödül alındığı yaşı hesaplayalım
df_filtered["age_at_award"] = df_filtered["year"] - df_filtered["birth_date"].dt.year

# Eksik veya geçersiz yaş değerlerini çıkartalım
df_filtered = df_filtered.dropna(subset=["age_at_award"])
df_filtered = df_filtered[df_filtered["age_at_award"] > 0]


plt.figure(figsize=(10, 6))
sns.swarmplot(data=df_filtered, x='category', y='age_at_award', palette='viridis')
plt.title('2000 Sonrası Nobel Ödülü Alanların Yaş Dağılımı Kategoriye Göre (Swarm Plot)')
plt.xlabel('Kategori')
plt.ylabel('Ödül Aldıklarında Yaş')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
